```python
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
```

### Filtering Data

```python
# getting data where 'Type 1' = Grass
df.loc[df['Type 1'] == 'Grass']
```

```python
# getting data by applying multiple filters
# get data where Type 1 = Grass and Type 2 = Poison

df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
```

```python
# getting data by applying multiple filters
# get data where Type 1 = Grass or Type 2 = Poison

df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')]
```

```python
# some more complex queries

new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
new_df
# you see that the index is not right so you just have to reset it...
new_df.reset_index(drop=True, inplace=True)
new_df

# this will still give old index to completely remove that
# new_df = new_df.reset_index(drop=True)
# new_df

new_df.to_csv('filtered.csv', index=False)
```

```python
# Now let's say we want to filter out all the pokemon whose name
# contains 'mega' in it...
df.loc[(df['Name'].str.contains('Mega'))][:5]
```

```python
# Now let's say we want to filter out all the pokemon whose name
# doesnot contains 'mega' in it...
df.loc[(~df['Name'].str.contains('Mega'))][:5]
```

```python
# We can also apply filtering using regular expressions

import re

df.loc[(df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True))]
# flags=re.I means ignore cases

# get all the pokemon whose name starts with Pi
df.loc[(df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True))]
```

### Conditional Changes

```python
# making all the 'Fire' in Type 1 to 'Flamer'
df.loc[(df['Type 1'] == 'Flamer', 'Type 1')] = 'Fire'
df
```

```python
# updating all the legendary pokemon to True where it is fire

df.loc[(df['Type 1'] == 'Fire', 'Legendary')] = True
df
```
