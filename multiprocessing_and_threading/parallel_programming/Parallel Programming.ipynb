# Parallel Programming in Python

Parallel computing is a type of computation in which many calculations or the execution of processes are carried out simultaniously. Large problems can often be divided into smaller ones, which can then be solved at the same time.



You have a problem and you realize that the problem can be subdivided into 4 parts, where each part doesn't have to know about the other parts, in such a case you can execute each part independent of each other, you can run them simultaniously.

Where if you have multiple processors, you basically assign one processor to each part of the problem. The obvious advantage of time.
time T can be divided into T/4 (4 processor)

**NOTE**
There are various ways to implement parallel computing, and there two ways to implement that in Python
- Multiprocessing
- Multithreading
