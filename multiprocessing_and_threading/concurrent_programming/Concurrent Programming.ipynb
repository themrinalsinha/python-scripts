# Concurrent Programming

## What is concurrent ?
Concurrent describes things that are occuring, or people who are doing something, at the same time, such as "concurrent users" of a computer program.

On the other hand, Consecutive refers to things that are arranged or happen in a sequential order.

**Example**
> Sequential Engineering vs. concurrent design & manufacturing
  

## What is concurrent programming/computing ?
- In a concurrent program, several streams of operations may execute concurrently. Each stream of operations executes as it would in a sequential program except for the fact that streams can communicate and interfere with one another.

- Concurrency is a property of a system which enables overlapping of process lifetimes.


We are triggering all the task at the same time, The switch of task from task1 to task2 and so on is called context switching. The lifetime of all these tasks are overlapped, this is concurrency.

Another thing that confuses alot of people is concurrency vs. parallelism

## Concurrency vs Parallelism
